```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("Diabetes_dataset.csv")
```

```python
x = data.iloc[:,:-1].values
x
```

```python
y = data.iloc[:,-1].values
y
```

```python
# le = LabelEncoder()
# y = le.fit_transform(y)
# y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
print("Training Set", x_train)
print("Testing Set", x_test)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
x_train
```

```python
ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=6, activation="relu"))
ann.add(tf.keras.layers.Dense(units=6, activation="relu"))
ann.add(tf.keras.layers.Dense(units=6, activation="relu"))
ann.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
ann.compile(optimizer="adam", loss="binary_crossentropy", metrics=['accuracy'])
```

```python
ann.fit(x_train, y_train, batch_size=5, epochs=100)
```

```python
# x_test
predicted = ann.predict(x_test)
pred = list(map(lambda predicted: 1 if predicted >= 0.5 else 0, predicted))
print("Predicted Values: ", predicted)
print("Predicted Values in Binary: ", pred)
print("Actual Values: ", y_test)
print("Accuracy: ", accuracy_score(y_test, pred))
```
